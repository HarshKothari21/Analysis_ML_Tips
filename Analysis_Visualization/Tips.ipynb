# Visualizing Data Tricks

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import re
```

## 1. Count Plot for counting value 0-1


```python
df = pd.read_csv('../../../machine learning/kaggle/Titanic/train.csv') 
```

```python
sns.countplot(df['Survived'])
```

## 2. Count plot as Value_counts

#### Works for lesser number of unique values only

```python
sns.countplot(df['Sex'])
```

## 3. Countplot for top 5 unique value

```python
df.Age.value_counts().head().plot(kind='bar')
```

## 4. CountPlot to compare two columns

```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant', 'rabbit', 'giraffe', 'coyote', 'horse']
df1 = pd.DataFrame({'speed': speed, 'lifespan': lifespan}, index=index)
ax = df1.plot.bar(rot=0)
```

## 5. Countplot to show attribute group-by another attribute

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
```

## 1. Show null values in dataset

```python
sns.heatmap(df.isnull(), yticklabels=False, cmap='plasma')
```

## 1. Look into relationships among dataset

```python
sns.heatmap(df.corr(), annot=True)
```

# 2. Configure the heatmap and figure size

```python
plt.subplots(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.2f');
```

# 3. Visualize only useful features

```python
corr = df.corr()
plt.subplots(figsize=(8, 5))
sns.heatmap(corr[((corr >= 0.3) | (corr <= -0.3)) & (corr != 1)], annot=True, linewidths=.5, fmt= '.2f');
```

# 1. Exploring Relationships between Predictors and Response
## Can also be use for Regression in machine learning 

```python
advert = pd.read_csv('../../../machine learning/Coursera_project/Sklearn/Linear_Regression/Advertising.csv')
```

```python
sns.pairplot(advert, x_vars=['TV', 'radio', 'newspaper'], y_vars='sales', height=7, aspect=0.7, kind='reg')
```

**From the graph we can conclude that TV and sales are more related where as newspaper and sales is very less or not at all related**
