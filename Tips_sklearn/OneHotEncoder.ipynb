## One Hot Encoder

```python
import pandas as pd
```

```python
df = pd.DataFrame ({'country' : ['India','U.S','Australia','India','Australia','India','U.S'],
'Age' : [44,34,28,27,30,42,25],
'Salary' : [72000,44000,35000,27000,32000,56000,45000],
'Purchased' : ['yes','no','yes','yes','no','yes','no']
})
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()

df1 = pd.DataFrame(encoder.fit_transform(df[['country']]).toarray())
```

```python
df = df.join(df1)
df
```

## Alternative approach

```python
df = pd.DataFrame ({'country' : ['India','U.S','Australia','India','Australia','India','U.S'],
'Age' : [44,34,28,27,30,42,25],
'Salary' : [72000,44000,35000,27000,32000,56000,45000],
'Purchased' : ['yes','no','yes','yes','no','yes','no']
})
```

```python
df.head()
```

```python
dum_df = pd.get_dummies(df.country, columns=["country"], prefix="is" )

df = df.join(dum_df)
df
```
